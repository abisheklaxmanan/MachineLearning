```python
import itertools
import numpy as np
import gzip
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import preprocessing
%matplotlib inline
import os
```

```python
# reading in dataset
training_data = pd.read_csv('GSE68086_TEP_data_matrix.txt', sep='\t').dropna()
training_data.head(57736)
```

```python
training_data = training_data.rename(columns={'Unnamed: 0': 'Gene Id'})
```

```python
training_data.shape
```

```python
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
```

Data uploaded, now we have to put the classifier.

```python
# features
params = training_data['Gene Id']
#creating labelEncoder
le = preprocessing.LabelEncoder()
# Converting string labels into numbers.
params_encoded=le.fit_transform(params)
# zip our feautures to our training data
zipped_features = list(list(zip(training_data[col].to_list(), params_encoded)) for col in training_data if col !='Unnamed: 0')
# assign data to labels  0 = cancer not present, 1 = cancer present 
labels = [(training_data[col].to_list(), 0) if 'WT' in col else  (training_data[col].to_list(), 1) for col in training_data.columns if col!= 'Unnamed: 0']
```

```python
X = [label[0] for label in labels]
Y = [label[1] for label in labels]
```

```python
# Split the data for cross-validation
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.33, random_state=seed)
```

```python
# set n for nearest neighbour 
n_1 = 1
n_49 = 49 
neigh_1 = KNeighborsClassifier(n_neighbors=n_1)
neigh_49 = KNeighborsClassifier(n_neighbors=n_49)
neigh_1.fit(X_train, Y_train)
neigh_49.fit(X_train, Y_train)
```

```python
Y_1_test_pred = neigh_1.predict(X_test)
metrics.roc_auc_score(Y_test, Y_1_test_pred)
```
